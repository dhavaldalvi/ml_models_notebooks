# Principal Component Analysis

We start with $p$-dimensional vectors, and want to summarize them by projecting down into a $q$-dimensional subspace. Also we assume that the data have been “centered”, so that every variable has mean 0. If we write the centered data in a matrix $\mathbf{X}$, where rows are objects and columns are variables, then $\mathbf{X}^{T}\mathbf{X}$ = $n\mathbf{C}$ ($(\mathbf{X}-\bar{\mathbf{X}})^{T}(\mathbf{X}-\bar{\mathbf{X}})$ = $n\mathbf{C}$ if not centered), where $\mathbf{C}$ is the covariance matrix of the data. We’ll start by looking for a one-dimensional projection. That is, we have $p$-dimensional vectors, and we want to project them on to a line, specified by unit vector $\mathbf{W}$, through the origin. The projection of data vector $\mathbf{X}_{i}$ on the line is $\mathbf{X}_{i}\cdot\mathbf{W}$ and the projection vector is $(\mathbf{X}_{i}\cdot\mathbf{W})\mathbf{W}$.

The mean of the projections will be zero, because the mean of the vectors $X_{i}$ is zero:
$$
\dfrac{1}{n}\sum_{i=1}^{n} \left(\mathbf{X}_{i}\cdot\mathbf{W}\right)\mathbf{W} = \left(\left(\dfrac{1}{n}\sum_{i=1}^{n} \mathbf{X}_{i}\right) \cdot \mathbf{W}\right) \mathbf{W} = 0
$$
If we use projected vector instead pf original vector there will be error because they do not coincide, for a vector $X_{i}$
$$
\begin{align*}
||X_{i} - (\mathbf{X}_{i}\cdot\mathbf{W})\mathbf{W}||^{2} &= (X_{i} - (\mathbf{X}_{i}\cdot\mathbf{W})\cdot\mathbf{W})(X_{i} - (\mathbf{X}_{i}\cdot\mathbf{W})\mathbf{W})\\
&= \mathbf{X}_{i}\cdot\mathbf{X}_{i} - \mathbf{X}_{i}\cdot(\mathbf{W}\cdot\mathbf{X}_{i})\mathbf{W} - (\mathbf{W}\cdot\mathbf{X}_{i})\mathbf{W}\cdot\mathbf{X}_{i} + (\mathbf{W}\cdot\mathbf{X}_{i})\mathbf{W}\cdot(\mathbf{W}\cdot\mathbf{X}_{i})\mathbf{W}\\
&= ||\mathbf{X}_{i}||^{2} - 2(\mathbf{W}\cdot\mathbf{X}_{i})^{2} + (\mathbf{W}\cdot\mathbf{X}_{i})^{2}\mathbf{W}\cdot\mathbf{W} \\
&= \mathbf{X}_{i}\cdot\mathbf{X}_{i} - (\mathbf{W}\cdot\mathbf{X}_{i})^{2} \quad\quad \text{since}\quad \mathbf{W}\cdot\mathbf{W} = ||\mathbf{W}||^{2} = 1
\end{align*}
$$
Adding the errors,
$$
\begin{align*}
\text{MSE}(\mathbf{W}) &= \dfrac{1}{n}\sum_{i=1}^{n} \left(||\mathbf{X}_{i}||^{2} - (\mathbf{W}\cdot\mathbf{X}_{i})^{2}\right)\\
&= \dfrac{1}{n}\left(\sum_{i=1}^{n}||\mathbf{X}_{i}||^{2} - \sum_{i=1}^{n}(\mathbf{W}\cdot\mathbf{X}_{i})^{2}\right)
\end{align*}
$$
The first summation doesn’t depend on $\mathbf{W}$. To make the MSE small, what we must do is make the second sum big, i.e., we want to maximize
$$
\dfrac{1}{n}\sum_{i=1}^{n}(\mathbf{W}\cdot\mathbf{X}_{i})^{2}
$$
The mean of a square is always equal to the square of the mean plus the variance:
$$
\dfrac{1}{n}\sum_{i=1}^{n}(\mathbf{W}\cdot\mathbf{X}_{i})^{2} = \left(\dfrac{1}{n}\sum_{i=1}^{n}\mathbf{W}\cdot\mathbf{X}_{i}\right)^{2} + \text{Var}(\mathbf{W}\cdot\mathbf{X}_{i})
$$
Since we’ve just seen that the mean of the projections is zero, minimizing the residual sum of squares turns out to be equivalent to maximizing the variance of the projections.

Of course in general we don’t want to project on to just one vector, but on to multiple principal components. If those components are orthogonal and have the uniy vectors $\mathbf{W}_{1}, \mathbf{W}_{2}\cdots \mathbf{W}_{j}$, then the image of $\mathbf{X}_{i}$ is its projection into the space spanned unit vectors $\mathbf{W}$ by these vectors,
$$
\sum_{j=1}^{k}(\mathbf{W}_{j}\cdot\mathbf{X}_{i})\mathbf{W}_{j}
$$
The mean of the projection on to each component is still zero. If we go through the same algebra for the mean squared error, it turns out that the cross-terms between different components all cancel out, and we are left with trying to maximize the sum of the variances of the projections on to the components.

If we stack our $n$ data vectors into an $n \times p$ matrix, $\mathbf{X}$, then the projections are given by $\mathbf{X}\mathbf{W}$, which is an $n \times 1$ matrix. The variance
$$
\begin{align*}
\text{Var}(\mathbf{W}\cdot\mathbf{X}_{i}) &= \dfrac{1}{n}\sum_{i}(\mathbf{X}_{i}\cdot\mathbf{W})^{2}\\
&= \dfrac{1}{n}(\mathbf{X}\mathbf{W})^{T}(\mathbf{X}\mathbf{W})\\
&= \dfrac{1}{n}\mathbf{W}^{T}\mathbf{X}^{T}\mathbf{X}\mathbf{W}\\
&= \mathbf{W}^{T}\dfrac{\mathbf{X}^{T}\mathbf{X}}{n}\mathbf{W}\\
&= \mathbf{W}^{T}\mathbf{C}\mathbf{W}
\end{align*}
$$
We have to maximise this with condition that $\mathbf{W}^{T}\mathbf{W} = 1$, by using lagrange multiplier method, we have
$$
\begin{align*}
L(\mathbf{W}, \lambda) &= \mathbf{W}^{T}\mathbf{C}\mathbf{W} - \lambda(\mathbf{W}^{T}\mathbf{W} - 1)\\
\dfrac{\partial L}{\partial \lambda} &= \mathbf{W}^{T}\mathbf{W} - 1 = 0\\
\dfrac{\partial L}{\partial \mathbf{W}} &= 2\mathbf{C}\mathbf{W} - 2\lambda\mathbf{W} = 0
\end{align*}
$$
Solving we get
$$
\begin{align*}
\mathbf{W}^{T}\mathbf{W} &= 1\\
\mathbf{C}\mathbf{W} &= \lambda\mathbf{W}
\end{align*}
$$
Consider, the second equation. Multiplying $\mathbf{W}^{T}$ from left on both side
$$
\text{Var}(\mathbf{W}\cdot\mathbf{X}_{i}) = \mathbf{W}^{T}\mathbf{C}\mathbf{W} = \lambda\mathbf{W}^{T}\mathbf{W} = \lambda
$$
Thus, desired vector $\mathbf{W}$ is an eigenvector of the covariance matrix $\mathbf{C}$, and the maximizing vector will be the one associated with the largest eigenvalue $\lambda$.

We know that $\mathbf{C}$ is a $p \times p$ matrix, so it will have $p$ different eigenvectors. Also we know that $\mathbf{C}$ is a covariance matrix, so it is symmetric, and then linear algebra tells us that the eigenvectors must be orthogonal to one another. Again because $\mathbf{C}$ is a covariance matrix, it is a positive matrix, in the sense that $\mathbf{X}\cdot\mathbf{C}\mathbf{X} \geq 0$ for any $\mathbf{X}$. This tells us that the eigenvalues of $\mathbf{C}$ must all be $\geq 0$. 

The eigenvectors of $\mathbf{C}$ ($\mathbf{W}$'s) are the principal components of the data.

For large dataset, computing covariance matrix is computationally tedious. Therefore, Single Value Decomposition (SVD) is used to get the principle components.

From SVD we have,
$$
\mathbf{X} = \mathbf{U}\Sigma \mathbf{V}^{T}
$$
then
$$
\mathbf{X}^{T}\mathbf{X} = (\mathbf{U}\Sigma \mathbf{V}^{T})^{T}(\mathbf{U}\Sigma \mathbf{V}^{T}) = \mathbf{V}\Sigma^{T}\mathbf{U}^{T}\mathbf{U}\Sigma \mathbf{V}^{T}
$$
since $\mathbf{U}^{T}\mathbf{U} = 1$ and $\Sigma^{T}\Sigma = \Sigma^{2} = \Lambda$, where $\Lambda$ is the matrix of eigenvalue placed diagonally
$$
\mathbf{X}^{T}\mathbf{X} = \mathbf{V}\Lambda\mathbf{V}^{T}
$$
which is (Eigenvalue Decomposition) EVD of covariance matrix $\mathbf{X}^{T}\mathbf{X}$. So, SVD calculates the same eigenvector matrix $\mathbf{V}$, without calculating covariance matrix.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple 2x2 matrix
A = np.array([[3, 1], [1, 3]])

# Perform Eigenvalue Decomposition
eigenvalues, eigenvectors = np.linalg.eig(A)

# Perform Singular Value Decomposition
U, S, VT = np.linalg.svd(A)

# Create vectors for visualization
origin = np.zeros((2,))  # origin point for quiver plots

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].set_title("Eigenvalue Decomposition")
ax[1].set_title("Singular Value Decomposition")

# Eigenvectors and their scaling by eigenvalues
for i in range(2):
    ax[0].quiver(*origin, *eigenvectors[:, i] * eigenvalues[i], angles='xy', scale_units='xy', scale=1, color=f"C{i}")
    ax[0].text(*(eigenvectors[:, i] * eigenvalues[i]), f"$\\lambda_{i+1}$", fontsize=12)

# SVD vectors (VT contains right singular vectors)
for i in range(2):
    ax[1].quiver(*origin, *(VT[i, :] * S[i]), angles='xy', scale_units='xy', scale=1, color=f"C{i}")
    ax[1].text(*(VT[i, :] * S[i]), f"$\\sigma_{i+1}$", fontsize=12)

# Common formatting
for a in ax:
    a.set_xlim(-5, 5)
    a.set_ylim(-5, 5)
    a.axhline(0, color='gray', lw=1)
    a.axvline(0, color='gray', lw=1)
    a.set_aspect('equal')
    a.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

# Generate synthetic 2D data
X, _ = make_blobs(n_samples=100, centers=1, cluster_std=2.5, random_state=42)
X = StandardScaler().fit_transform(X)  # Center and scale data

# Method 1: PCA via Eigenvalue Decomposition
# Compute covariance matrix
cov_matrix = np.cov(X.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_matrix)

# Sort by largest eigenvalues
sorted_idx = np.argsort(eigen_vals)[::-1]
eigen_vals = eigen_vals[sorted_idx]
eigen_vecs = eigen_vecs[:, sorted_idx]

# Method 2: PCA via SVD
U, S, VT = np.linalg.svd(X, full_matrices=False)
svd_components = VT.T  # Right singular vectors

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
titles = ["PCA via Eigenvalue Decomposition", "PCA via SVD"]

for i, a in enumerate(ax):
    a.scatter(X[:, 0], X[:, 1], alpha=0.5)
    a.set_aspect('equal')
    a.axhline(0, color='gray', lw=1)
    a.axvline(0, color='gray', lw=1)
    a.set_xlim(-4, 4)
    a.set_ylim(-4, 4)
    a.set_title(titles[i])
    
    # Plot the principal components
    if i == 0:
        for j in range(2):
            vec = eigen_vecs[:, j] * 3  # Scale for visibility
            a.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=f"C{j}", label=f"PC{j+1}")
    else:
        for j in range(2):
            vec = svd_components[:, j] * 3
            a.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=f"C{j}", label=f"PC{j+1}")

    a.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate a small 2D dataset (easier to visualize)
np.random.seed(42)
X = np.random.randn(10, 2) @ np.array([[2, 1], [1, 2]])  # Stretch and rotate
X = X - np.mean(X, axis=0)  # Center the data

# --- PCA via EVD ---
# Step 1: Covariance matrix
cov_matrix = np.cov(X.T)

# Step 2: Eigen decomposition
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors
sorted_idx = np.argsort(eig_vals)[::-1]
eig_vals = eig_vals[sorted_idx]
eig_vecs = eig_vecs[:, sorted_idx]

# Project data onto top 1 principal component
X_evd_proj = X @ eig_vecs[:, :1]

# --- PCA via SVD ---
U, S, VT = np.linalg.svd(X, full_matrices=False)
X_svd_proj = X @ VT.T[:, :1]  # Project onto first principal component

# Plotting original and projected data
fig, axs = plt.subplots(1, 3, figsize=(16, 5))

# Original data
axs[0].scatter(X[:, 0], X[:, 1], c='gray', label='Original Data')
axs[0].quiver(0, 0, eig_vecs[0, 0]*3, eig_vecs[1, 0]*3, angles='xy', scale_units='xy', scale=1, color='r', label='PC1')
axs[0].quiver(0, 0, eig_vecs[0, 1]*3, eig_vecs[1, 1]*3, angles='xy', scale_units='xy', scale=1, color='b', label='PC2')
axs[0].set_title("Original Data with PCs (via EVD)")
axs[0].axis('equal')
axs[0].legend()

# Projection via EVD
axs[1].scatter(X_evd_proj, np.zeros_like(X_evd_proj), c='r')
axs[1].set_title("Projection via EVD")
axs[1].axhline(0, color='gray', lw=1)
axs[1].axis('equal')

# Projection via SVD
axs[2].scatter(X_svd_proj, np.zeros_like(X_svd_proj), c='b')
axs[2].set_title("Projection via SVD")
axs[2].axhline(0, color='gray', lw=1)
axs[2].axis('equal')

plt.tight_layout()
plt.show()
```

```python
class PCA:
    def __init__(self, n_components = None, method='svd'):
        self.n_components = n_components
        self.method = method
        self.eigen_vectors = None
        self.eigen_values = None
    
    def fit(self, X):
        X = (X - np.mean(X, axis=0))/np.std(X)
        if self.method == 'svd':

            # Singular Value Decomposition
            U, S, self.VT = np.linalg.svd(X, full_matrices=False)

            self.eigen_vectors = self.VT

            self.eigen_values = 2*(S**2)/(X.size - 1)

            # Project onto first principal component
            # X_svd_proj = X @ VT.T[:, :1]

        elif self.method == 'evd':

            # Covariance matrix
            cov_matrix = np.cov(X.T)

            # Eigen decomposition
            eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

            # Sort eigenvalues and eigenvectors
            sorted_idx = np.argsort(eig_vals)[::-1]
            eig_vals = eig_vals[sorted_idx]
            self.eig_vecs = eig_vecs[:, sorted_idx]

            self.eigen_vectors = self.eig_vecs

            self.eigen_values = eig_vals

            # Project data onto top 1 principal component
            # X_evd_proj = X @ eig_vecs[:, :1]

    def transform(self, X):
        if self.method == 'svd':
            X_svd_proj = X @ self.VT.T[:, :self.n_components]
            return X_svd_proj

        elif self.method == 'evd':
            X_evd_proj = X @ self.eig_vecs[:, :self.n_components]
            return X_evd_proj
```
